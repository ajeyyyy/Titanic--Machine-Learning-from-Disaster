### 2.1 Filling Mising Values

#### 2.1.1 Age

Here we can see that the <b>Age</b> and <b>Cabin</b> feature have significant amount of missing values.

To fill Age column I am computing a Title column from "Name". Which i will use to fill the median <b>age</b> value of each title group in the missing field.

```python
import re

titleRegex = re.compile(r' ([A-Za-z]+\.)')

title = lambda x: titleRegex.search(x).groups()[0]

train_test['Title'] = train_test['Name'].apply(title)
```

```python
train_test.head()
```

```python
train_test['Title'].value_counts()
```

We can see that Mr, Miss, Mrs and Master title have some significant count in the dataset and others are not that considerable. So we can map the title as follows:
- Mr : 0  
- Miss : 1  
- Mrs: 2  
- Master: 3
- Others: 4

```python
title_mapping = {"Mr.": 0, "Miss.": 1, "Mrs.": 2, "Master.": 3, "Dr.": 4, "Rev.": 4, "Col.": 4, "Major.": 4, "Mlle.": 4,
                "Ms.": 3, "Lady.": 4, "Jonkheer.": 4, "Mme.": 4,"Capt.": 4,"Countess.": 4, "Sir.": 4, "Don.": 4, "Dona." : 4 }

train_test['Title'] = train_test.Title.map(title_mapping)
train_test.head()
```

```python
train_test['Age'].fillna(train_test.groupby('Title')['Age'].transform('median'), inplace = True)
train_test.head()
```

```python
train_test.isnull().sum()/len(train_test) * 100
```

Mapping Sex values to integers

```python
sex_mapping = {"male": 0, "female": 1}

train_test['Sex'] = train_test.Sex.map(sex_mapping)
train_test.head()
```

#### 2.1.2 Cabin

Now lets process <b>cabin</b> feature

```python
train_test['Cabin']
```

As you can see do not need cabin numbers to evaluate the survival, we can just group them all by just extracting the first letter from their numbers.

```python
cabin = lambda x: x[0] if pd.notnull(x) else 'X'

train_test['Cabin'] = train_test['Cabin'].apply(cabin)
train_test.head()
```

```python
train_test.isnull().sum()/len(train_test) * 100
```

Now just <b>Embarked</b> and <b>Fare</b> are left. Obviously the Survived feature is which is null is from Test.csv so we can omit that.

#### 2.1.3 Fare

For fare we can just fill the median value of the that column grouped by Pclass.

```python
train_test['Fare'].fillna(train_test.groupby('Pclass')['Fare'].transform('median'), inplace=True)
train_test.head()
```

```python
train_test.isnull().sum()/len(train_test) * 100
```

#### 2.1.4 Embarked

Since there are only few missing value in <b>Embarked</b> feature we can just the value whic appears the maximum number of times.

```python
train_test.groupby('Embarked').count()
```

As you can see many people are emabarked from <b>'S'</b>, so it is highly probable that the missing values might be 'S' too.

```python
train_test['Embarked'].fillna('S', inplace = True)
```

```python
train_test.isnull().sum()/len(train_test) * 100
```

Now we have no missing values

Before moving on lets drop useless features <b>PassengerId</b> and <b>Name</b>.

```python
train_test.drop(labels = ['PassengerId','Name'], axis = 1, inplace = True)
train_test.head()
```

#### 2.2.6 Ticket

The ticket number may convey the details about where the passenger have been placed on the ship which may also actually contribute to survival probablity. It may show which cabins the passenger has been placed or it may also which class the passenger belong to as well.
- We will just remove the ticket prefixes and replace it with ticket feacture.

```python
ticketPrefixRegex = re.compile(r'^([A-Za-z].*) ')

ticketPrefix = lambda x: ticketPrefixRegex.search(x).groups()[0].replace(".","").replace("/","").replace(" ","") if ticketPrefixRegex.search(x) else 'X'

train_test['Ticket'] = train_test['Ticket'].apply(ticketPrefix)
```

```python
train_test.Ticket.unique()
```

As you can see it is a cateogorical value we will just create dummy varialbles for it too.

```python
train_test = pd.get_dummies(train_test, columns = ['Ticket'], prefix="T")
train_test.head()
```

I am going to compare the following classification models and evaluate the mean accuracy for each of them using stratified kfold(Cross validation).

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
cv = StratifiedKFold(n_splits=10, random_state=0, shuffle=True)
```

### 3.1 Logistic Regression

```python
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression(solver='liblinear')
accuracy_logreg = cross_val_score(logreg_model, X, y, cv=cv, scoring='accuracy')

print("Cross Validation Scores for Log Reg: {}".format(accuracy_logreg))
print("Mean Accuracy Score for Log Reg: {}".format(np.mean(accuracy_logreg)))
```

```python
logreg_model.fit(X,y)
survived = logreg_model.predict(test)
```
